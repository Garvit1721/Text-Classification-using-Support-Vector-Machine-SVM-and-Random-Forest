```python
from nltk.stem import WordNetLemmatizer
from nltk.corpus import wordnet
from nltk import pos_tag
```

```python
lemmatizer = WordNetLemmatizer()
```

```python
from nltk.corpus import stopwords
import string
stops = set(stopwords.words('english'))
punc = list(string.punctuation)
stops.update(punc)
stops = list(stops)
stops.append('flight')
stops.append('airline')
stops.append('flights')
stops.append('AA')
stops.remove('@')
stops
```

```python
def get_simple_pos(tag):
    if tag.startswith("J"):
        return wordnet.ADJ
    if tag.startswith("V"):
        return wordnet.VERB
    if tag.startswith("N"):
        return wordnet.NOUN
    if tag.startswith("R"):
        return wordnet.ADV
    else :
        return wordnet.NOUN
```

```python
def clean_review(words):
    output_word = []
    for w in words :
        if w.lower() not in stops :
            pos = pos_tag([w])
            clean_word = lemmatizer.lemmatize(w,pos=get_simple_pos(pos[0][1]))
            output_word.append(clean_word.lower())
    return output_word
```

```python
documents = [(clean_review(document),category) for document,category in word_train]
```

```python
documents[0]
```

```python
train_doc = [" ".join(doc) for doc,category in documents]
```

```python
train_doc
```

```python
import emoji
import re
def clean_text(text):
    text_no_emojis = ''.join(c for c in text if not emoji.is_emoji(c))
    text_no_extra_spaces = ' '.join(text_no_emojis.split())
    text_no_links = re.sub(r'http\s*//\S+', '', text_no_extra_spaces)
    text_no_links = re.sub('((www\.[^\s]+)|(https?://[^\s]+))','',text_no_links) 
    text_no_links = re.sub('@ [^\s]+','',text_no_links)
    text_no_links = re.sub('[\s]+', ' ', text_no_links)
    text_no_links = re.sub(r'#([^\s]+)', r'\1', text_no_links) 
    text_no_links = text_no_links.strip('\'"')
    text_no_numbers = ''.join(c for c in text_no_links if not c.isnumeric())
    return text_no_numbers
train_doc = [clean_text(s) for s in train_doc]
```
